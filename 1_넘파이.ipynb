### **â€» ì„ í˜• ëŒ€ìˆ˜**

ì„ í˜• ëŒ€ìˆ˜ëŠ” ë²¡í„°ì™€ í–‰ë ¬ì„ ì‚¬ìš©í•´ ê³µê°„ê³¼ ë³€í™˜ì„ ë‹¤ë£¨ëŠ” ìˆ˜í•™ì˜ í•œ ë¶„ì•¼ì…ë‹ˆë‹¤. ì‰½ê²Œ ë§í•´, ì—¬ëŸ¬ ìˆ«ìë¥¼ ì²´ê³„ì ìœ¼ë¡œ ë°°ì—´í•œ ë²¡í„°(ë¦¬ìŠ¤íŠ¸ ê°™ì€ ê²ƒ)ì™€ í–‰ë ¬(ìˆ«ìê°€ ê²©ìë¡œ ë°°ì—´ëœ í‘œ)ì„ ì´ìš©í•´ ë°ì´í„°ë¥¼ í‘œí˜„í•˜ê³  ì¡°ì‘í•˜ëŠ” ë°©ë²•ì…ë‹ˆë‹¤. ì„ í˜• ëŒ€ìˆ˜ëŠ” ì»´í“¨í„° ê·¸ë˜í”½ìŠ¤, ë¨¸ì‹ ëŸ¬ë‹, ë¬¼ë¦¬í•™, ê³µí•™ ë“± ë‹¤ì–‘í•œ ë¶„ì•¼ì—ì„œ ì‚¬ìš©ë©ë‹ˆë‹¤. ì˜ˆë¥¼ ë“¤ì–´, ì‚¬ì§„ì—ì„œ ìƒ‰ìƒì„ ì¡°ì •í•˜ê±°ë‚˜, ë¡œë´‡ì´ ì›€ì§ì¼ ê²½ë¡œë¥¼ ê³„ì‚°í•˜ê±°ë‚˜, AIê°€ ë°ì´í„°ë¥¼ ë¶„ì„í•˜ëŠ” ë°ë„ ì„ í˜• ëŒ€ìˆ˜ê°€ í•„ìš”í•©ë‹ˆë‹¤.

## **5. ì •ë ¬**

NumPyì˜Â np.sort()Â í•¨ìˆ˜ëŠ”Â ë°°ì—´ì„Â ì •ë ¬í•˜ëŠ”Â ë°Â ì‚¬ìš©ë©ë‹ˆë‹¤.Â ì´Â í•¨ìˆ˜ëŠ”Â ê¸°ë³¸ì ìœ¼ë¡œÂ ì›ë˜Â ë°°ì—´ì„Â ë³€ê²½í•˜ì§€Â ì•Šê³ Â ì •ë ¬ëœÂ ë°°ì—´ì˜Â ë³µì‚¬ë³¸ì„Â ë°˜í™˜í•©ë‹ˆë‹¤.

````python
ndarr1 = np.array([1, 10, 5, 7, 2, 4, 3, 6, 8, 9])
print(ndarr1)

print(np.sort(ndarr1)) # ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬
print(ndarr1)

print(np.sort(ndarr1)[::-1]) # ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
````

````python
ndarr2d = np.array([[11, 10, 12, 9],
                    [3, 1, 4, 2],
                    [5, 6, 7, 8]])
print(ndarr2d.shape)

# ì—´ ë‹¨ìœ„ë¡œ ì •ë ¬. ê° ì—´ì˜ ê°’ë“¤ì„ ê°œë³„ì ìœ¼ë¡œ ì˜¤ë¦„ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•˜ë©°, í–‰(row)ì˜ ìˆœì„œëŠ” ì¬ë°°ì—´
np.sort(ndarr2d, axis=0)
# í–‰(row) ë‹¨ìœ„ë¡œ ì •ë ¬. ê° í–‰ì˜ ì›ì†Œë“¤ì´ ê°œë³„ì ìœ¼ë¡œ ì˜¤ë¦„ì°¨ìˆœìœ¼ë¡œ ì •ë ¬
np.sort(ndarr2d, axis=1)
# í–‰ ë‹¨ìœ„ë¡œ ì •ë ¬í•œ í›„, ê° í–‰ì˜ ê°’ì„ ì—­ìˆœìœ¼ë¡œ ì •ë ¬í•˜ëŠ” ë™ì‘ì„ ìˆ˜í–‰
np.sort(ndarr2d, axis=1)[:, ::-1]
# axis=1ê³¼ ë™ì¼í•˜ê²Œ ë™ì‘. -1ì€ ë°°ì—´ì˜ ë§ˆì§€ë§‰ ì¶•ì„ ê¸°ì¤€ìœ¼ë¡œ ë™ì‘í•˜ë¼ëŠ” ì˜ë¯¸. 2ì°¨ì› ë°°ì—´ì—ì„œ ë§ˆì§€ë§‰ ì¶•ì€ axis=1ì„ ë‚˜íƒ€ëƒ„
np.sort(ndarr2d, axis=-1)
````
